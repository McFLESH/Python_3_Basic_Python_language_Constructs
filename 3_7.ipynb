```python
# Необходимо написать программу, которая проверяет, является ли введенное число простым.

a = int(input())
if a > 1:
  for i in range(2, a - 1):
    if a % i == 0:
      print('Число не является простым')
      break
    else: i = i + 1
if i == a - 1:
  print('Число является простым')
```
