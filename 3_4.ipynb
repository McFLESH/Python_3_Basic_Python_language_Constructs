```python
# На вход подаётся число N. Вывести сумму всех чисел от 1 до N.

a = int(input())
i = 0
sum = 0
while i < a:
  i = i + 1
  sum = sum + i
print("Сумма чисел от 1 до ", a," = ", sum)
```
