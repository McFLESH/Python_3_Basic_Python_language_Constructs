```python
# Требуется написать программу,
# которая запрашивает у пользователя его возраст и выводит соответствующую возрастную катеорию
v = int(input())
if v < 18:
  print('детский')
elif 18 < v < 29:
  print('молодой')
elif 30 < v < 59:
  print('взрослый')
else:
  print('пожилой')
```
