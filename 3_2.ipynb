```python
# Напишите программу, которая считывает со стандартного ввода целые числа,
# по одному числу в строке, и после первого введенного нуля выводит сумму полученных на вход чисел.

summa = 0
while True:
  i = int(input())
  if i == 0:
    break
  else:
    summa = summa + i
print(summa)
```
