```python
# Написать программу, которая принимает на входе 2 слова – логин и пароль.
# Если логин равен «student» и пароль равен «0000», то вывести сообщение «Вход разрешен!», иначе вывести сообщение «Ошибка. Вход запрещен»

l = input('введите логин: ')
p = input('введите пароль: ')
if (l == 'student') and (p == '0000'):
  print('Вход разрешен!')
else: print('Ошибка. Вход запрещен')
```
